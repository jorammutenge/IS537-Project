```python
import gensim
from gensim import corpora
from gensim.parsing.preprocessing import remove_stopwords
import nltk
from nltk.corpus import stopwords
from nltk import pos_tag
# nltk.download('averaged_perceptron_tagger')

# func to add topics
def add_topic_labels(df):
    # create a list of documents from the 'script' column
    documents = list(df['script'])

    # define stop words, tags to remove, and words to remove
    stop_words = stopwords.words('english')
    removal = ['ADV','PRON','CCONJ','PUNCT','PART','DET','ADP','SPACE', 'NUM', 'SYM']
    remove_words = ['like','gone','know','right','na','gon','yeah','really','okay','even','sure','miller','lisa',
                    'get','gonna','well','thank','oh','uh','hi','got','um','go','little','mortimer','mayfield',
                    'would','great','come','hey','wanna','hmm','mr','yes','good','give','jorgy','coleman','ruth',
                    'going','want','let','think','one','us','look','make','ed','gotta','ben','dale','randolph'
                    'something','back','see','need','man','say','sorry','could','madoff','wrong','eric','robert',
                    'thing','lt','lf','made','way','said','sir','two','new','naomi','ha','ok','around','jean','ron',
                    'louis','winthorpe','gekko','gordon','buddy','jordan','tell','erin','sam','eric','seen','michael','seth',
                    'patrick','paul','allen','listen','god','bateman','marcus','heh','bernie','big','steve','jimmy',
                    'anything','christie','patrick','dorsia','garfield','mark','maybe','nick','donnie','name','went'
                   ]

    # remove stop words and unwanted words
    tokenized_docs = [[token for token in gensim.utils.simple_preprocess(doc) if token not in stop_words and token not in remove_words] for doc in documents]

    # remove tags
    tagged_docs = [nltk.pos_tag(doc) for doc in tokenized_docs]
    tokenized_docs = [[token for token, pos in doc if pos not in removal] for doc in tagged_docs]

    # create a dictionary from the tokenized documents
    dictionary = corpora.Dictionary(tokenized_docs)

    # create a bag-of-words representation of the documents
    corpus = [dictionary.doc2bow(doc) for doc in tokenized_docs]

    # train an LDA model on the corpus
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=dictionary,
                                                num_topics=5,
                                                random_state=42,
                                                passes=10)

    # extract the top fifteen words for each topic
    topic_labels = []
    for doc in tokenized_docs:
        bow = dictionary.doc2bow(doc)
        topics = lda_model.get_document_topics(bow)
        top_fifteen_words = [dictionary[word] for word, prob in sorted(lda_model.get_topic_terms(max(topics, key=lambda x: x[1])[0], topn=15), key=lambda x: x[1], reverse=True)]
        topic_labels.append(top_fifteen_words)

    # add the topic labels as a new column in the dataframe
    df['topic'] = topic_labels
    return df
```

```python
# create mapping to create dataframe for each movie
movie_dict = {'Margin_Call':0, 'Wall_Street':1, 'The_Big_Short':2, 'Trading_Places':3, 'The_Wolf_of_Wall_Street':4,'American_Psycho':5, 
              'Arbitrage':6, 'Equity':7, 'Inside_Job':8, 'Boiler_Room':9, 'Rogue_Trader':10, 'The_Wizard_of_Lies':11, 'Other_Peoples_Money':12
             }

# loop through dictionary to create dataframes
for key, value in movie_dict.items():
    df_key = pd.DataFrame([[df.loc[value, 'movie'], df.loc[value, 'script']]], columns=['movie', 'script']).pipe(add_topic_labels)
    globals()[key] = df_key
```

```python
# List of dataframes
dataframes = [Margin_Call, Wall_Street, The_Big_Short, Trading_Places, The_Wolf_of_Wall_Street, American_Psycho,
              Arbitrage, Equity, Inside_Job, Boiler_Room, Rogue_Trader, The_Wizard_of_Lies, Other_Peoples_Money]

# Initialize an empty dataframe to store the concatenated data
movie_df = pd.DataFrame()

# Concatenate the dataframes vertically
for df in dataframes:
    movie_df = pd.concat([movie_df, df], axis=0)

# Reset the index of the concatenated dataframe
movie_df = movie_df.reset_index(drop=True)
```

```python
#func to plot topics
def plot_word_grid(words, suptitle):
    # Define the figure size and background color
    fig = plt.figure(figsize=(8, 4), facecolor=BLACK, dpi=DPI)

    # Set the title
    fig.suptitle(suptitle, fontproperties=FUTURA, color='white', fontsize=15, y=.98)

    # Define the layout of the subplots
    layout = (5, 5)

    # Create subplots and add words to them
    for i, word in enumerate(words):
        ax = fig.add_subplot(*layout, i+1)
        ax.set_facecolor(BLACK)
        ax.text(0.5, 0.5, word, fontsize=10, ha='center', va='center', fontproperties=PAYPAL, color=YELLOW)

        # Remove spines and tick labels
        for side in 'top,right,bottom,left'.split(','):
            ax.spines[side].set_visible(False)
        ax.tick_params(axis='both', which='both', length=0)
        ax.set_xticklabels([])
        ax.set_yticklabels([])

    # Adjust the spacing between subplots
    fig.subplots_adjust(hspace=0.2, wspace=0.2)
    
    # Show the figure
    plt.show()
```

## Profanity
Too much profanity in these movies. Let's try to investigate it.

```python
# lists of bad words
f_words = ['fuck', 'fucked', 'fucking', 'fuckin']
s_words = ['shit', 'shitting']

(movie_df
 .assign(f_word=lambda df_: df_.script.str.count('|'.join(f_words)),
         s_word=lambda df_: df_.script.str.count('|'.join(s_words)))
)
```

```python
(movie_df
 .assign(f_word=lambda df_: df_.script.str.count('|'.join(f_words)),
         s_word=lambda df_: df_.script.str.count('|'.join(s_words)))
 [['movie', 'f_word', 's_word']]
 .set_index('movie')
 .rename_axis(None, axis=0)
)
```

```python
# func to normalize values
def normalize_df(df):
    df['f_word'] = (df['f_word'] - df['f_word'].min()) / (df['f_word'].max() - df['f_word'].min())
    df['s_word'] = (df['s_word'] - df['s_word'].min()) / (df['s_word'].max() - df['s_word'].min())
    return df
```

```python
(movie_df
 .assign(f_word=lambda df_: df_.script.str.count('|'.join(f_words)),
         s_word=lambda df_: df_.script.str.count('|'.join(s_words)))
 [['movie', 'f_word', 's_word']]
 .set_index('movie')
 .rename_axis(None, axis=0)
 .pipe(normalize_df)
)
```

```python
profane = (movie_df
 .assign(f_word=lambda df_: df_.script.str.count('|'.join(f_words)),
         s_word=lambda df_: df_.script.str.count('|'.join(s_words)))
 [['movie', 'f_word', 's_word']]
 .set_index('movie')
 .rename_axis(None, axis=0)
 .pipe(normalize_df)
)

fig, ax = plt.subplots(figsize=(8, 3.5), dpi=DPI, facecolor=BLACK)
ax.set_facecolor(BLACK)
(profane
 .plot.scatter('s_word', 'f_word', ax=ax, color=YELLOW, marker='h', s=22)
)

my_index = ['The Wolf of Wall Street', 'The Big Short', 'The Wizard of Lies', 'Boiler Room', 'Rogue Trader']
for i, row in profane.iterrows():
    if i in my_index:
        ax.text(row['s_word'], row['f_word'], str(i), color=WHITE, fontproperties=PAYPAL, fontsize=8)
fig.suptitle('Profanity in Wall Street movies', fontproperties=FUTURA, c=WHITE, fontsize=18, y=.92)
ax.set_ylabel('F-words', color=WHITE, fontproperties=PAYPAL, fontsize=10)
ax.set_xlabel('S-words', color=WHITE, fontproperties=PAYPAL, fontsize=10)
# ax.tick_params(bottom=False)
ax.tick_params(axis='x', colors=YELLOW)
ax.tick_params(axis='y', colors=YELLOW)
for tick in ax.get_xticklabels():
    tick.set_fontproperties(PAYPAL)
for tick in ax.get_yticklabels():
    tick.set_fontproperties(PAYPAL)
for side in 'top,right'.split(','):
    ax.spines[side].set_visible(False)
for side in 'bottom,left'.split(','):
    ax.spines[side].set_color(WHITE)
```
